# **YOLO**

```python
import time
import os
import numpy as np
import torch
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
from ultralytics import YOLO
from PIL import Image, ImageDraw, ImageFont

class YOLOv8Benchmark:
    def __init__(self, model_path, test_folder, batch_sizes=[1], device=None):
        """
        Khởi tạo benchmark cho model YOLOv8

        Args:
            model_path (str): Đường dẫn đến file trọng số
            test_folder (str): Đường dẫn đến thư mục chứa ảnh test
            batch_sizes (list): Danh sách các batch size để test
            device (str): Device sử dụng ('cpu', 'cuda', 'mps' hoặc None để tự chọn)
        """
        self.model_path = model_path
        self.test_folder = test_folder
        self.batch_sizes = batch_sizes

        # Tự động chọn device nếu không chỉ định
        if device is None:
            self.device = 'cuda' if torch.cuda.is_available() else 'cpu'
        else:
            self.device = device

        # Load model
        self.model = self._load_model()

        # Lấy danh sách ảnh
        self.image_files = self._get_image_files()

        print(f"Model đã được tải với device: {self.device}")
        print(f"Tổng số ảnh test: {len(self.image_files)}")

    def _load_model(self):
        """Load model YOLOv8 với device phù hợp"""
        try:
            model = YOLO(self.model_path)
            return model
        except Exception as e:
            print(f"Lỗi khi tải model: {e}")
            return None

    def _get_image_files(self):
        """Lấy danh sách các file ảnh trong thư mục test"""
        image_extensions = ('.jpg', '.jpeg', '.png', '.bmp', '.tif', '.tiff')
        if os.path.exists(self.test_folder):
            return [os.path.join(self.test_folder, f) for f in os.listdir(self.test_folder)
                    if f.lower().endswith(image_extensions)]
        else:
            print(f"Thư mục {self.test_folder} không tồn tại")
            return []

    def benchmark_single(self, warmup=3, repeats=5):
        """
        Đo hiệu suất inference với từng ảnh riêng lẻ

        Args:
            warmup (int): Số lần chạy làm nóng trước khi đo
            repeats (int): Số lần lặp lại mỗi ảnh để tính trung bình

        Returns:
            dict: Kết quả benchmark
        """
        print(f"Đang thực hiện benchmark cho {len(self.image_files)} ảnh (warmup={warmup}, repeats={repeats})...")

        results = {
            'image_name': [],
            'inference_time': [],
            'num_objects': [],
            'image_size': []
        }

        # Warmup
        if warmup > 0:
            print("Đang làm nóng model...")
            for _ in range(warmup):
                if len(self.image_files) > 0:
                    self.model(self.image_files[0])

        # Benchmark từng ảnh
        for img_path in tqdm(self.image_files, desc="Đang xử lý ảnh"):
            file_name = os.path.basename(img_path)

            # Lấy kích thước ảnh
            with Image.open(img_path) as img:
                img_size = f"{img.width}x{img.height}"

            # Chạy inference nhiều lần và tính trung bình
            times = []
            detections = None

            for _ in range(repeats):
                torch.cuda.synchronize() if self.device == 'cuda' else None
                start = time.time()
                detection_results = self.model(img_path)
                torch.cuda.synchronize() if self.device == 'cuda' else None
                end = time.time()
                times.append(end - start)
                detections = detection_results

            # Tính thời gian trung bình
            avg_time = np.mean(times)

            # Đếm số object phát hiện được
            num_objects = len(detections[0].boxes)

            # Lưu kết quả
            results['image_name'].append(file_name)
            results['inference_time'].append(avg_time)
            results['num_objects'].append(num_objects)
            results['image_size'].append(img_size)

        # Tính thống kê tổng hợp
        if results['inference_time']:
            avg_inference_time = np.mean(results['inference_time'])
            min_inference_time = np.min(results['inference_time'])
            max_inference_time = np.max(results['inference_time'])
            std_inference_time = np.std(results['inference_time'])

            print(f"\nKết quả benchmark single image:")
            print(f"- Tổng số ảnh: {len(self.image_files)}")
            print(f"- Thời gian inference trung bình: {avg_inference_time:.4f} giây")
            print(f"- Thời gian inference nhỏ nhất: {min_inference_time:.4f} giây")
            print(f"- Thời gian inference lớn nhất: {max_inference_time:.4f} giây")
            print(f"- Độ lệch chuẩn: {std_inference_time:.4f} giây")
            print(f"- FPS trung bình: {1/avg_inference_time:.2f}")

        return pd.DataFrame(results)

    def benchmark_batch(self):
        """
        Đo hiệu suất với các batch size khác nhau

        Returns:
            dict: Kết quả benchmark theo batch size
        """
        batch_results = {
            'batch_size': [],
            'total_time': [],
            'avg_time_per_image': [],
            'fps': []
        }

        print(f"Đang thực hiện benchmark cho các batch size: {self.batch_sizes}")

        for batch_size in self.batch_sizes:
            # Tạo các batch ảnh
            batches = [self.image_files[i:i + batch_size]
                      for i in range(0, len(self.image_files), batch_size)]

            batch_start_time = time.time()

            # Xử lý từng batch
            for batch in tqdm(batches, desc=f"Xử lý batch size {batch_size}"):
                torch.cuda.synchronize() if self.device == 'cuda' else None
                self.model(batch)
                torch.cuda.synchronize() if self.device == 'cuda' else None

            batch_end_time = time.time()
            total_time = batch_end_time - batch_start_time
            avg_time = total_time / len(self.image_files)
            fps = 1 / avg_time

            batch_results['batch_size'].append(batch_size)
            batch_results['total_time'].append(total_time)
            batch_results['avg_time_per_image'].append(avg_time)
            batch_results['fps'].append(fps)

            print(f"Batch size {batch_size}:")
            print(f"- Tổng thời gian: {total_time:.4f} giây")
            print(f"- Thời gian trung bình mỗi ảnh: {avg_time:.4f} giây")
            print(f"- FPS: {fps:.2f}")

        return pd.DataFrame(batch_results)

    def visualize_results(self, single_results, batch_results):
        """Hiển thị kết quả benchmark bằng đồ thị"""
        plt.figure(figsize=(15, 10))

        # Đồ thị thời gian inference theo từng ảnh
        plt.subplot(2, 2, 1)
        plt.bar(single_results['image_name'], single_results['inference_time'], color='skyblue')
        plt.title('Thời gian inference theo từng ảnh')
        plt.xlabel('Tên ảnh')
        plt.ylabel('Thời gian (giây)')
        plt.xticks(rotation=90, fontsize=8)
        plt.tight_layout()

        # Đồ thị thời gian inference theo số object
        plt.subplot(2, 2, 2)
        plt.scatter(single_results['num_objects'], single_results['inference_time'], alpha=0.7)
        plt.title('Thời gian inference theo số object')
        plt.xlabel('Số object phát hiện')
        plt.ylabel('Thời gian (giây)')
        plt.grid(True, linestyle='--', alpha=0.7)

        # Đồ thị batch size và FPS
        plt.subplot(2, 2, 3)
        plt.plot(batch_results['batch_size'], batch_results['fps'], 'o-', color='green')
        plt.title('FPS theo batch size')
        plt.xlabel('Batch size')
        plt.ylabel('FPS')
        plt.grid(True, linestyle='--', alpha=0.7)

        # Đồ thị batch size và thời gian trung bình
        plt.subplot(2, 2, 4)
        plt.plot(batch_results['batch_size'], batch_results['avg_time_per_image'], 'o-', color='red')
        plt.title('Thời gian xử lý trung bình theo batch size')
        plt.xlabel('Batch size')
        plt.ylabel('Thời gian mỗi ảnh (giây)')
        plt.grid(True, linestyle='--', alpha=0.7)

        plt.tight_layout()
        plt.show()

    def save_visualized_detections(self, output_folder):
        """Lưu kết quả phát hiện với bounding box"""
        os.makedirs(output_folder, exist_ok=True)

        print(f"Đang lưu kết quả phát hiện vào thư mục {output_folder}")

        for img_path in tqdm(self.image_files, desc="Đang lưu kết quả phát hiện"):
            filename = os.path.basename(img_path)
            results = self.model(img_path)

            # Lưu ảnh với bounding box
            result_img = results[0].plot()
            output_path = os.path.join(output_folder, f"detection_{filename}")
            Image.fromarray(result_img).save(output_path)

        print(f"Đã lưu {len(self.image_files)} ảnh với kết quả phát hiện")

def main():
    # Các thông số
    model_path = '/content/Weight/best.pt'  # Đường dẫn đến model
    test_folder = '/content/test/images'    # Thư mục chứa ảnh test
    output_folder = './detection_results'   # Thư mục lưu kết quả
    batch_sizes = [1, 2, 4, 8]              # Các batch size để test

    # Khởi tạo đối tượng benchmark
    yolo_benchmark = YOLOv8Benchmark(
        model_path=model_path,
        test_folder=test_folder,
        batch_sizes=batch_sizes
    )

    # Benchmark với từng ảnh riêng lẻ
    single_results = yolo_benchmark.benchmark_single(warmup=3, repeats=3)

    # Benchmark với các batch size khác nhau
    batch_results = yolo_benchmark.benchmark_batch()

    # Hiển thị kết quả bằng đồ thị
    yolo_benchmark.visualize_results(single_results, batch_results)

    # Lưu ảnh với bounding box
    yolo_benchmark.save_visualized_detections(output_folder)

    # Xuất kết quả ra file CSV
    single_results.to_csv("single_image_results.csv", index=False)
    batch_results.to_csv("batch_results.csv", index=False)

    print("Hoàn tất benchmark!")

if __name__ == "__main__":
    main()
```

# **Faster R-CNN**

```python
import time
import os
import numpy as np
import torch
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
import cv2

# Các import của Detectron2
import detectron2
from detectron2.config import get_cfg
from detectron2.model_zoo import model_zoo
from detectron2.engine import DefaultPredictor
from detectron2.utils.visualizer import Visualizer
from detectron2.data import MetadataCatalog
import detectron2.data.transforms as T


class FasterRCNNBenchmarkNoConfig:
    def __init__(self, weights_path, test_folder, batch_sizes=[1], device=None, num_classes=None):
        """
        Khởi tạo benchmark cho model Faster R-CNN của Detectron2 mà không cần file cấu hình gốc

        Args:
            weights_path (str): Đường dẫn đến file trọng số
            test_folder (str): Đường dẫn đến thư mục chứa ảnh test
            batch_sizes (list): Danh sách các batch size để test
            device (str): Device sử dụng ('cpu', 'cuda' hoặc None để tự chọn)
            num_classes (int): Số lượng lớp trong mô hình (nếu biết)
        """
        self.weights_path = weights_path
        self.test_folder = test_folder
        self.batch_sizes = batch_sizes
        self.num_classes = num_classes

        # Tự động chọn device nếu không chỉ định
        if device is None:
            self.device = 'cuda' if torch.cuda.is_available() else 'cpu'
        else:
            self.device = device

        # Tạo cấu hình và tải model
        self.cfg = self._create_config()
        self.predictor = self._load_model()

        # Cấu hình transform
        self.transform_gen = self._setup_transforms()

        # Lấy danh sách ảnh
        self.image_files = self._get_image_files()

        print(f"Model đã được tải với device: {self.device}")
        print(f"Tổng số ảnh test: {len(self.image_files)}")

    def _create_config(self):
        """Tạo cấu hình mới dựa trên cấu hình mặc định của Detectron2"""
        cfg = get_cfg()

        # Sử dụng cấu hình mặc định của Faster R-CNN với backbone ResNet50
        # Bạn có thể thay đổi thành cấu hình khác nếu biết mô hình của mình dựa trên kiến trúc nào
        cfg.merge_from_file(model_zoo.get_config_file("COCO-Detection/faster_rcnn_R_101_FPN_3x.yaml"))

        # Cấu hình device
        cfg.MODEL.DEVICE = self.device

        # Đường dẫn đến weights
        cfg.MODEL.WEIGHTS = self.weights_path

        # Ngưỡng confidence
        cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.5

        # Đặt số lớp nếu được cung cấp
        if self.num_classes is not None:
            cfg.MODEL.ROI_HEADS.NUM_CLASSES = self.num_classes

        return cfg

    def _setup_transforms(self):
        """Thiết lập các transform cho ảnh đầu vào"""
        # Tạo một instance của ResizeShortestEdge
        resize_transform = T.ResizeShortestEdge(
            [self.cfg.INPUT.MIN_SIZE_TEST, self.cfg.INPUT.MIN_SIZE_TEST],
            self.cfg.INPUT.MAX_SIZE_TEST
        )

        # Trả về transform (không sử dụng TransformList)
        return resize_transform

    def _load_model(self):
        """Load model Faster R-CNN của Detectron2"""
        try:
            predictor = DefaultPredictor(self.cfg)
            return predictor
        except Exception as e:
            print(f"Lỗi khi tải model: {e}")
            return None

    def _get_image_files(self):
        """Lấy danh sách các file ảnh trong thư mục test"""
        image_extensions = ('.jpg', '.jpeg', '.png', '.bmp', '.tif', '.tiff')
        if os.path.exists(self.test_folder):
            return [os.path.join(self.test_folder, f) for f in os.listdir(self.test_folder)
                    if f.lower().endswith(image_extensions)]
        else:
            print(f"Thư mục {self.test_folder} không tồn tại")
            return []

    def benchmark_single(self, warmup=3, repeats=5):
        """
        Đo hiệu suất inference với từng ảnh riêng lẻ

        Args:
            warmup (int): Số lần chạy làm nóng trước khi đo
            repeats (int): Số lần lặp lại mỗi ảnh để tính trung bình

        Returns:
            dict: Kết quả benchmark
        """
        print(f"Đang thực hiện benchmark cho {len(self.image_files)} ảnh (warmup={warmup}, repeats={repeats})...")

        results = {
            'image_name': [],
            'inference_time': [],
            'num_objects': [],
            'image_size': []
        }

        # Warmup
        if warmup > 0 and len(self.image_files) > 0:
            print("Đang làm nóng model...")
            warmup_img = cv2.imread(self.image_files[0])
            for _ in range(warmup):
                self.predictor(warmup_img)

        # Benchmark từng ảnh
        for img_path in tqdm(self.image_files, desc="Đang xử lý ảnh"):
            file_name = os.path.basename(img_path)

            # Đọc ảnh
            img = cv2.imread(img_path)

            # Lấy kích thước ảnh
            img_height, img_width = img.shape[:2]
            img_size = f"{img_width}x{img_height}"

            # Chạy inference nhiều lần và tính trung bình
            times = []
            outputs = None

            for _ in range(repeats):
                torch.cuda.synchronize() if self.device == 'cuda' else None
                start = time.time()
                outputs = self.predictor(img)
                torch.cuda.synchronize() if self.device == 'cuda' else None
                end = time.time()
                times.append(end - start)

            # Tính thời gian trung bình
            avg_time = np.mean(times)

            # Đếm số object phát hiện được (instances)
            if "instances" in outputs:
                num_objects = len(outputs["instances"])
            else:
                num_objects = 0

            # Lưu kết quả
            results['image_name'].append(file_name)
            results['inference_time'].append(avg_time)
            results['num_objects'].append(num_objects)
            results['image_size'].append(img_size)

        # Tính thống kê tổng hợp
        if results['inference_time']:
            avg_inference_time = np.mean(results['inference_time'])
            min_inference_time = np.min(results['inference_time'])
            max_inference_time = np.max(results['inference_time'])
            std_inference_time = np.std(results['inference_time'])

            print(f"\nKết quả benchmark single image:")
            print(f"- Tổng số ảnh: {len(self.image_files)}")
            print(f"- Thời gian inference trung bình: {avg_inference_time:.4f} giây")
            print(f"- Thời gian inference nhỏ nhất: {min_inference_time:.4f} giây")
            print(f"- Thời gian inference lớn nhất: {max_inference_time:.4f} giây")
            print(f"- Độ lệch chuẩn: {std_inference_time:.4f} giây")
            print(f"- FPS trung bình: {1/avg_inference_time:.2f}")

        return pd.DataFrame(results)

    def benchmark_batch(self):
        """
        Đo hiệu suất với các batch size khác nhau

        Lưu ý: Detectron2 không hỗ trợ batch inference trực tiếp trong DefaultPredictor,
        nên chúng ta sẽ mô phỏng bằng cách xử lý tuần tự và tính toán hiệu suất

        Returns:
            dict: Kết quả benchmark theo batch size
        """
        batch_results = {
            'batch_size': [],
            'total_time': [],
            'avg_time_per_image': [],
            'fps': []
        }

        print(f"Đang thực hiện benchmark cho các batch size: {self.batch_sizes}")

        for batch_size in self.batch_sizes:
            # Tạo các batch ảnh
            batches = [self.image_files[i:i + batch_size]
                      for i in range(0, len(self.image_files), batch_size)]

            batch_start_time = time.time()

            # Xử lý từng batch
            for batch in tqdm(batches, desc=f"Xử lý batch size {batch_size}"):
                # Xử lý từng ảnh trong batch
                for img_path in batch:
                    img = cv2.imread(img_path)
                    torch.cuda.synchronize() if self.device == 'cuda' else None
                    self.predictor(img)
                    torch.cuda.synchronize() if self.device == 'cuda' else None

            batch_end_time = time.time()
            total_time = batch_end_time - batch_start_time
            avg_time = total_time / len(self.image_files)
            fps = 1 / avg_time

            batch_results['batch_size'].append(batch_size)
            batch_results['total_time'].append(total_time)
            batch_results['avg_time_per_image'].append(avg_time)
            batch_results['fps'].append(fps)

            print(f"Batch size {batch_size}:")
            print(f"- Tổng thời gian: {total_time:.4f} giây")
            print(f"- Thời gian trung bình mỗi ảnh: {avg_time:.4f} giây")
            print(f"- FPS: {fps:.2f}")

        return pd.DataFrame(batch_results)

    def visualize_results(self, single_results, batch_results):
        """Hiển thị kết quả benchmark bằng đồ thị"""
        plt.figure(figsize=(15, 10))

        # Đồ thị thời gian inference theo từng ảnh
        plt.subplot(2, 2, 1)
        plt.bar(single_results['image_name'], single_results['inference_time'], color='skyblue')
        plt.title('Thời gian inference theo từng ảnh')
        plt.xlabel('Tên ảnh')
        plt.ylabel('Thời gian (giây)')
        plt.xticks(rotation=90, fontsize=8)
        plt.tight_layout()

        # Đồ thị thời gian inference theo số object
        plt.subplot(2, 2, 2)
        plt.scatter(single_results['num_objects'], single_results['inference_time'], alpha=0.7)
        plt.title('Thời gian inference theo số object')
        plt.xlabel('Số object phát hiện')
        plt.ylabel('Thời gian (giây)')
        plt.grid(True, linestyle='--', alpha=0.7)

        # Đồ thị batch size và FPS
        plt.subplot(2, 2, 3)
        plt.plot(batch_results['batch_size'], batch_results['fps'], 'o-', color='green')
        plt.title('FPS theo batch size')
        plt.xlabel('Batch size')
        plt.ylabel('FPS')
        plt.grid(True, linestyle='--', alpha=0.7)

        # Đồ thị batch size và thời gian trung bình
        plt.subplot(2, 2, 4)
        plt.plot(batch_results['batch_size'], batch_results['avg_time_per_image'], 'o-', color='red')
        plt.title('Thời gian xử lý trung bình theo batch size')
        plt.xlabel('Batch size')
        plt.ylabel('Thời gian mỗi ảnh (giây)')
        plt.grid(True, linestyle='--', alpha=0.7)

        plt.tight_layout()
        plt.savefig("faster_rcnn_benchmark_results.png", dpi=300, bbox_inches='tight')
        plt.show()

    def save_visualized_detections(self, output_folder):
        """Lưu kết quả phát hiện với bounding box"""
        os.makedirs(output_folder, exist_ok=True)

        # Lấy metadata (sử dụng metadata mặc định nếu không có sẵn)
        metadata = MetadataCatalog.get(self.cfg.DATASETS.TEST[0] if len(self.cfg.DATASETS.TEST) else "__unused")

        print(f"Đang lưu kết quả phát hiện vào thư mục {output_folder}")

        for img_path in tqdm(self.image_files, desc="Đang lưu kết quả phát hiện"):
            filename = os.path.basename(img_path)

            # Đọc ảnh
            img = cv2.imread(img_path)

            # Thực hiện inference
            outputs = self.predictor(img)

            # Tạo visualizer
            v = Visualizer(img[:, :, ::-1], metadata=metadata, scale=1.0)

            # Vẽ các kết quả phát hiện
            if "instances" in outputs:
                instances = outputs["instances"]
                vis_output = v.draw_instance_predictions(instances.to("cpu"))
                result_img = vis_output.get_image()[:, :, ::-1]  # Convert RGB to BGR

                # Lưu ảnh
                output_path = os.path.join(output_folder, f"detection_{filename}")
                cv2.imwrite(output_path, result_img)

        print(f"Đã lưu {len(self.image_files)} ảnh với kết quả phát hiện")


def main():
    # Các thông số
    weights_path = '/content/best.pth'  # Đường dẫn đến file trọng số
    test_folder = '/content/test/images'              # Thư mục chứa ảnh test
    output_folder = './detection_results'             # Thư mục lưu kết quả
    batch_sizes = [1, 2, 4, 8]                        # Các batch size để test
    num_classes = 80  # Đặt số lượng lớp của mô hình nếu biết (ví dụ: 80 cho COCO)

    # Khởi tạo đối tượng benchmark
    faster_rcnn_benchmark = FasterRCNNBenchmarkNoConfig(
        weights_path=weights_path,
        test_folder=test_folder,
        batch_sizes=batch_sizes,
        num_classes=num_classes
    )

    # Benchmark với từng ảnh riêng lẻ
    single_results = faster_rcnn_benchmark.benchmark_single(warmup=3, repeats=3)

    # Benchmark với các batch size khác nhau
    batch_results = faster_rcnn_benchmark.benchmark_batch()

    # Hiển thị kết quả bằng đồ thị
    faster_rcnn_benchmark.visualize_results(single_results, batch_results)

    # Lưu ảnh với bounding box
    faster_rcnn_benchmark.save_visualized_detections(output_folder)

    # Xuất kết quả ra file CSV
    single_results.to_csv("single_image_results_faster_rcnn.csv", index=False)
    batch_results.to_csv("batch_results_faster_rcnn.csv", index=False)

    print("Hoàn tất benchmark!")

if __name__ == "__main__":
    main()
```
